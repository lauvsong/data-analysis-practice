```python
fig, axes = plt.subplots(2,2)
for i in range(2):
  for ii in range(2):
    axes[i,ii].plot(np.random.randn(50*(i+1)*(ii+1)).cumsum(), "k-")

plt.show()
```

```python
fig, axes = plt.subplots(2,2)
for i in range(2):
  for ii in range(2):
    axes[i,ii].plot(np.random.randn(50*(i+1)*(ii+1)).cumsum(), "k-")

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)

for i in range(2):
  for ii in range(2):
    axes[i,ii].plot(np.random.randn(50*(i+1)*(ii+1)).cumsum(), "k-")

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(arr1)
ax.plot(arr2)
ax.set_title('arr1 & arr2')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(arr1)
ax.plot(arr2)
ax.set_title('arr1 & arr2')

ticks = ax.set_xticks([0,100,200,300,400,500,700,800,1000])
labels = ax.set_xticklabels(['zero', 'one', 'two', 'three', 'four'],
                            rotation=20,
                            fontsize='medium')
ax.set_xlabel('Stages')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(arr1, 'r-', label='arr1')
ax.plot(arr1, 'b.', label='arr2')
ax.set_title('arr1 & arr2')
ax.legend(loc='best')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(arr1, 'r--', label='num1')
ax.plot(arr2, 'b--', label='num2')

ax.text(500,0,
        "(500,0)",
        family="monospace",
        fontsize=10)

plt.show()
```
